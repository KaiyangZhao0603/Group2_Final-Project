```python
import utils as u
import tensorflow_hub as hub

yamnet = hub.load('/Users/zhaokaiyang/.cache/kagglehub/models/google/yamnet/tensorFlow2/yamnet/1')
```

```python
# 指定你的音频文件夹路径
folder_path = 'FMA_Excerpt/'
# 处理文件夹并获取 embeddings
embeddings_dict = u.process_folder_FMA(folder_path, yamnet)
```

```python
# 打印所有键
print(list(embeddings_dict.keys()))
```

```python
# 创建一个新字典，其键是转换后的格式
new_embeddings_dict = {}
for key in embeddings_dict.keys():
    new_key = key[:-4] + ".0"  # 移除 '.mp3' 并添加 '.0'
    new_embeddings_dict[new_key] = embeddings_dict[key]

embeddings_dict = new_embeddings_dict
```

```python
# 打印所有键
print(list(embeddings_dict.keys()))
```

```python
from tensorflow.keras.models import load_model
from tcn import TCN  # 确保正确导入 TCN 层的定义

# 加载模型时指定自定义层
model_predict = load_model('/Users/zhaokaiyang/Documents/24Spring/Deep Learning/Final Project/emotionPredict.h5', custom_objects={'TCN': TCN})
```

```python
import numpy as np
# 假设你的模型期望的输入名为'tcn_input'
# 将所有embeddings整合为一个numpy数组（假设每个embedding的维度相同）
all_embeddings = np.array(list(embeddings_dict.values()))

# 用正确的键名创建新的字典来进行预测
input_dict = {'tcn_input': all_embeddings}

# 进行预测
predicted_values = model_predict.predict(input_dict)
```

```python
predicted_values
```

```python
import pandas as pd

# 歌曲编号列表
song_ids = list(embeddings_dict.keys())
# 创建DataFrame
df = pd.DataFrame(predicted_values, columns=['Valence', 'Arousal'])
df['Song_ID'] = song_ids  # 添加歌曲编号为新的一列

# 调整列的顺序
df = df[['Song_ID', 'Valence', 'Arousal']]

### 3. 导出DataFrame到CSV文件
df.to_csv('predicted_emotions.csv', index=False)
```
