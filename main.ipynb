```python
import csv

data_home = '/Users/apple/Documents/DL/Final-project/PMEmo2019/chorus'
labelfile = '/Users/apple/Documents/DL/Final-project/PMEmo2019/annotations/static_annotations.csv'

# get the music id in label_ids
# and all label info in label_dict
label_ids = []
label_dict = {}
with open(labelfile, newline='') as csvfile:
    reader = csv.reader(csvfile)
    headers = next(reader) 
    for row in reader:
        label_ids.append(row[0])
        label_dict[row[0]]=[float(row[1]), float(row[2])]
```

```python
print(label_dict)
```

```python
from sklearn.model_selection import train_test_split

track_ids = label_ids

train_ids, test_ids = train_test_split(track_ids, test_size=0.2, random_state=124)
train_ids, val_ids = train_test_split(train_ids, test_size=0.25, random_state=124)
```

```python
print(train_ids)
```

```python
# Waveform
sample_rate = 16000  # 16kHz for YAMNet
augment = False
input_shape = (29*sample_rate,)

input_args_train = [data_home, augment, label_dict, train_ids, sample_rate]
input_args_val   = [data_home, augment, label_dict, val_ids, sample_rate]
input_args_test  = [data_home, augment, label_dict, test_ids, sample_rate]

dataset_train = u.create_dataset(u.wav_generator, input_args_train, input_shape)
dataset_val = u.create_dataset(u.wav_generator, input_args_val, input_shape)
dataset_test= u.create_dataset(u.wav_generator, input_args_test, input_shape)
```
